```python
import altair as alt
import pandas as pd
alt.data_transformers.disable_max_rows()
source = pd.read_csv("C:/Users/jackt/OneDrive/Desktop/JackLarson01.github.io/_data/League_Dataset.csv")
source = source.drop(['url', 'gameid', 'year', 'datacompleteness', 'participantid', 'playerid', 'playoffs', 'teamid'], axis = 1)
source = source.dropna(subset=['champion'])
source
```

```python
champs = source.loc[:, ['champion']]
bar = alt.Chart(champs).mark_bar().encode(
    x = alt.X("champion", title="Champion Name"),
    y = alt.Y("count()", title='Total Game Appearances')
).interactive().properties(
    width=3000,
    height=300
)
bar
```

```python
bar.properties(width='container').save("C:/Users/jackt/OneDrive/Desktop/JackLarson01.github.io/assets/json/League_appearance_bar.json")
```
